```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from minisom import MiniSom
```

```python
df=pd.read_csv('Credit_Card_Applications.csv')
```

```python
df.head()
```

```python
X=df.drop('Class',axis=1)
y=df['Class']
```

```python
X.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_scaled=scaler.fit_transform(X)
```

```python
som=MiniSom(x=10,y=10,input_len=15,sigma=1,learning_rate=0.5)  ##input_len, Number of the elements of the vectors in input
som.random_weights_init(X_scaled)
som.train_random(X_scaled,num_iteration=100)
```

```python
# Create a mapping of data points to SOM neurons
win_map = som.win_map(X_scaled)
```

```python
plt.figure(figsize=(10, 10))
plt.title('U-Matrix (Unified Distance Matrix)')
# weights = som.get_weights()
# plt.xlim(0, weights.shape[0])  # x-dimension is the first element of the shape
# plt.ylim(0, weights.shape[1])
# plt.imshow(som.distance_map().T,cmap='viridis', interpolation='nearest')  ##coolwarm
plt.imshow(som.distance_map().T, cmap='viridis', interpolation='nearest', origin='lower', extent=[0, som.get_weights().shape[0], 0, som.get_weights().shape[1]])
plt.colorbar(label='Distance')
plt.show()
```

```python
# Visualize Clusters on SOM Grid
plt.figure(figsize=(8, 8))
plt.title('Clusters on SOM Grid')
markers = ['o', 's']
colors = ['r', 'g']

for i, x in enumerate(X_scaled):
    w = som.winner(x)
    plt.plot(w[0] + 0.5,
         w[1] + 0.5,
         markers[y[i]],
         markeredgecolor = colors[y[i]],
         markerfacecolor = 'None',
         markersize = 10,
         markeredgewidth = 2)

# for i, x in enumerate(X_scaled):
#     w = som.winner(x)
#     plt.scatter(w[0]+0.5, 
#             w[1]+0.5, 
#             marker=markers[y[i]], 
#             edgecolor=colors[y[i]], 
#             facecolor='None', 
#             s=100, # Marker size
#             linewidths=2) # Marker edge width

# weights = som.get_weights()
# plt.xlim(0, weights.shape[0])  # x-dimension is the first element of the shape
# plt.ylim(0, weights.shape[1])
# plt.imshow(som.distance_map().T,cmap='coolwarm',  alpha=0.5, interpolation='nearest')  #cmap='coolwarm',
plt.imshow(som.distance_map().T, cmap='gray', interpolation='nearest', origin='lower', extent=[0, som.get_weights().shape[0], 0, som.get_weights().shape[1]])
plt.colorbar(label='Distance')
plt.show()
```

```python
## Finding the frauds

# Create a mapping of data points to SOM neurons
win_map = som.win_map(X_scaled)

frauds=np.concatenate((win_map[3,5],win_map[5,6]),axis=0)
```

```python
fraud_custom=scaler.inverse_transform(frauds)
```

```python
pd.DataFrame(fraud_custom,columns=['CustomerID', 'A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9',
       'A10', 'A11', 'A12', 'A13', 'A14'])
```

## Going from Unsupervised to Supervised Deep Learning

```python
##creating matrix of features
customers=df.iloc[:,1:]  ##remove the customer id column
```

```python
customers.shape
```

```python
is_fraud=np.zeros(len(df))
```

```python
is_fraud.shape
```

```python
df.iloc[1,0]
```

```python
fraud_custom[:,0]
```

```python
for i in range(len(df)):
    if df.iloc[i,0] in fraud_custom[:,0]:
        is_fraud[i]=1
```

```python
sum(is_fraud)
```

```python
## ANN

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
customers = sc.fit_transform(customers)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# early_stop=EarlyStopping(monitor='val_loss',patience=5)
```

```python
model=Sequential()
```

```python
model.add(Dense(2,activation='relu'))
# model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(customers,is_fraud,epochs=15) #,validation_split=0.2,callbacks=[early_stop])
```

```python
## Predicting the test set results
```

```python
y_pred=model.predict(customers)
```

```python
y_pred=(y_pred>.5).astype(int)
```
